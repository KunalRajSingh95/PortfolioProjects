```python
#Ordering Data
data_frame.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
#Scatter Plot with Budget vs Gross

plt.scatter(x=data_frame['budget'],y=data_frame['gross'])
plt.title("Budget Vs Gross Earnings")
plt.xlabel("Gross Earnings")
plt.ylabel("Budget For Film")
```

```python
#Budget Vs Gross using Seaborn

sns.regplot(x="budget",y="gross",data=data_frame,scatter_kws={"color":"red"},line_kws={"color":"purple"})
```

```python
#Company analysis Numeric Representation
data_frame_num = data_frame
for col_name in data_frame_num.columns:
    if(data_frame_num[col_name].dtype == 'object'):
        data_frame_num[col_name]=data_frame_num[col_name].astype('category')
        data_frame_num[col_name]=data_frame_num[col_name].cat.codes
data_frame_num.head()
```

```python
correlation_matrix = data_frame_num.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")
```

```python
correlation_mat = data_frame_num.corr()
corr_pairs = correlation_mat.unstack()
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr=sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
